Relative Strength Index

```python
from talib import (RSI, BBANDS, MACD,
                   NATR, WILLR, WMA,
                   EMA, SMA, CCI, CMO,
                   MACD, PPO, ROC,
                   ADOSC, ADX, MOM, STOCH, ULTOSC, MFI)
```

```python
for t in T:
    print(f'{t:02}_RSI')
```

```python
for t in T:
    stock_ohlc[f'{t:02}_RSI'] = RSI(stock_ohlc['Close'], timeperiod=t)
```

```python
len(stock_ohlc.columns)
```

```python
for t in T:
    stock_ohlc[f'{t:02}_WILLR'] =WILLR(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], timeperiod=t)
```

Compute Bollinger Bands

```python
for t in T:
    upperband, middleband, Lowerband = BBANDS(stock_ohlc['Close'], timeperiod=t, nbdevup=2, nbdevdn=2, matype=0)
    stock_ohlc[f'{t:02}_BBH'] = upperband
    stock_ohlc[f'{t:02}_BBM'] = middleband
    stock_ohlc[f'{t:02}_BBL'] = Lowerband
```

Se codidifica el algoritmo de labelizado descrito en el artículo citado en la memoria.

```python
# Se usa una ventana de 11 posiciones, prestando especial atención al punto medio

window_size = 11
half_window = window_size // 2

# Se inicializa una nueva columna para las etiquetas en stock_ohlcFrame
stock_ohlc['Label'] = 'Hold'  # Etiqueta por defecto
print(len(stock_ohlc))
# Restableciendo el índice del DataFrame
stock_ohlc.reset_index(drop=True, inplace=True)

# Ahora ejecuta el bucle con el código modificado
for i in range(half_window, len(stock_ohlc) - half_window):
    middle_price = stock_ohlc.at[i, 'Close']
    max_price = stock_ohlc['Close'][i - half_window:i + half_window + 1].max()
    min_price = stock_ohlc['Close'][i - half_window:i + half_window + 1].min()
    
    if middle_price == max_price:
        stock_ohlc.at[i, 'Label'] = 'Sell'
    elif middle_price == min_price:
        stock_ohlc.at[i, 'Label'] = 'Buy'



# Se muestra las primeras 30 filas
stock_ohlc[['Close', 'Label']].head(30)
len(stock_ohlc)

```

Se observa que se hayan creado correctamente las etiquetas

Vamos a guardar una copia de los datos

Primero se eliminan las columnas que no servirán para formar las imágenes, así como los registros con valores nulos.

```python
ohlcv_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
stock_ohlc = stock_ohlc.drop(ohlcv_columns, axis=1)
stock_ohlc = stock_ohlc.dropna()
stock_ohlc
```

```python
stock_ohlc.info()
```

Se resetea y se ordena el indice por precaución.

```python
stock_ohlc.reset_index(drop=True, inplace=True)
stock_ohlc.sort_index()
```

```python
stock_ohlc['Label']
```

Transformamos las etiquetas en valores numéricos de cara a posteriores procesamientos.

```python
stock_ohlc['Label'] = stock_ohlc['Label'].apply(lambda x: 0 if x == 'Hold' else (1 if x == 'Buy' else (2 if x == 'Sell' else x)))
```

Se guarda tanto las imágenes(features) como las labels(targets) en formato ".csv".

```python
stock_ohlc.drop('Label', axis=1).to_csv('features.csv',index = False)
stock_ohlc['Label'].to_csv('targets.csv', index = False)
```
