Se importan las lirería necesarias para el correcto manejo de los datos.

```python
import pandas as pd
import yfinance as yf
import pandas as pd
```

Se obtienen los datos históricos del índice sp500.

```python
# Define el símbolo de la acción que te interesa
ticker = '^SPX'  

# Obtén los datos históricos
df_stock = yf.download(ticker, start="2000-01-01", end="2023-12-31")

df_stock
```

Se elimina la columna 'Adj Close'.

```python
df_stock.drop(columns='Adj Close', inplace=True)
df_stock
```

```python
df_stock.info()
```

Se ordena el índice.

```python
df_stock.sort_index(inplace=True)
```

```python
df_stock
```

Se realiza una copia de los datos por seguridad.

```python
stock_ohlc = df_stock.copy()
```

Se generan los indicadores técnicos

```python
T = list(range(6, 21))
```

Relative Strength Index

```python
from talib import (RSI, BBANDS, MACD,
                   NATR, WILLR, WMA,
                   EMA, SMA, CCI, CMO,
                   MACD, PPO, ROC,
                   ADOSC, ADX, MOM, STOCH, ULTOSC, MFI)
```

```python
for t in T:
    print(f'{t:02}_RSI')
```

```python
for t in T:
    stock_ohlc[f'{t:02}_RSI'] = RSI(stock_ohlc['Close'], timeperiod=t)
```

```python
len(stock_ohlc.columns)
```

Williams %R

```python
for t in T:
    stock_ohlc[f'{t:02}_WILLR'] =WILLR(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], timeperiod=t)
```

```python
len(stock_ohlc.columns)
```

```python
import warnings

# Ignorar advertencias
warnings.filterwarnings('ignore')
```

Compute Bollinger Bands

```python
for t in T:
    upperband, middleband, Lowerband = BBANDS(stock_ohlc['Close'], timeperiod=t, nbdevup=2, nbdevdn=2, matype=0)
    stock_ohlc[f'{t:02}_BBH'] = upperband
    stock_ohlc[f'{t:02}_BBM'] = middleband
    stock_ohlc[f'{t:02}_BBL'] = Lowerband
```

```python
len(stock_ohlc.columns)
```

Normalized Average True Range

```python
for t in T:
    stock_ohlc[f'{t:02}_NATR'] = NATR(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], timeperiod=t)
```

```python
len(stock_ohlc.columns)
```

Percentage Price Oscillator

```python
for t in T:
    stock_ohlc[f'{t:02}_PPO'] = PPO(stock_ohlc['Close'], fastperiod=t, slowperiod=t*14, matype=0)
```

```python
len(stock_ohlc.columns)
```

Moving Average Convergence/Divergence

```python
for t in T:
    macd, macdsignal, macdhist = MACD(stock_ohlc['Close'], fastperiod=t, slowperiod=t+14, signalperiod=t-3)

    stock_ohlc[f'{t:02}_MACD'] = macd
    stock_ohlc[f'{t:02}_SIGNAL'] = macdsignal
    stock_ohlc[f'{t:02}_HISTOGRAM'] = macdhist
```

```python
len(stock_ohlc.columns)
```

Momentum

```python
for t in T:
    stock_ohlc[f'{t:02}_MOM'] = MOM(stock_ohlc['Close'], timeperiod=t)
```

Weighted Moving Average

```python
for t in T:
    stock_ohlc[f'{t:02}_WMA'] = WMA(stock_ohlc['Close'], timeperiod=t)
```

Exponential Moving Average

```python
for t in T:
    stock_ohlc[f'{t:02}_EMA'] = EMA(stock_ohlc['Close'], timeperiod=t)
```

Commodity Channel Index

```python
for t in T:    
    stock_ohlc[f'{t:02}_CCI'] = CCI(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], timeperiod=t)
```

Chande Momentum Oscillator

```python
for t in T:
    stock_ohlc[f'{t:02}_CMO'] = CMO(stock_ohlc['Close'], timeperiod=t)
```

Rate of Change

```python
for t in T:
    stock_ohlc[f'{t:02}_ROC'] = ROC(stock_ohlc['Close'], timeperiod=t)
```

Chaikin A/D Oscillator

```python
for t in T:
    stock_ohlc[f'{t:02}_ADOSC'] = ADOSC(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], stock_ohlc['Volume'], fastperiod=t-3, slowperiod=4+t)
```

Average Directional Movement Index

```python
for t in T:
    stock_ohlc[f'{t:02}_ADX'] = ADX(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], timeperiod=t)
```

```python
stock_ohlc
```

SMA

```python
for t in T:
    stock_ohlc[f'{t:02}_SMA'] = SMA(stock_ohlc['Close'], timeperiod=t)
```

Stochastic Oscillator

```python
for t in T:
    sLowk, sLowd = STOCH(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], 
                           fastk_period=t+8, slowk_period=t-3, 
                           slowk_matype=0, slowd_period=t-3, slowd_matype=0)

    stock_ohlc[f'{t:02}_slowk'] = sLowk
    stock_ohlc[f'{t:02}_slowd'] = sLowd
```

Ultimate Oscillator

```python
for t in T:
    stock_ohlc[f'{t:02}_ULTOSC'] = ULTOSC(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], 
                                timeperiod1=t+1, timeperiod2=t+8, timeperiod3=t+22)
```

MFI

```python
for t in T:
    stock_ohlc[f'{t:02}_MFI'] = MFI(stock_ohlc['High'], stock_ohlc['Low'], stock_ohlc['Close'], stock_ohlc['Volume'], timeperiod=t)
```

```python
stock_ohlc
```
